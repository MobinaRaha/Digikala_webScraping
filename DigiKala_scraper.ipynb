```python

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import psycopg2


import time
driver = webdriver.Chrome()

# Load the webpage
url = "https://www.digikala.com/search/category-mobile-phone/product-list/?sort=7"
driver.get(url)
```

```python
# Wait for the page to load completely (you might need to adjust the time based on your network speed)
driver.implicitly_wait(5)
# time.sleep(10)


# Track the number of elements loaded
prev_num_elements = 0

# Loop until we have at least 200 elements
while True:

    # Scroll to the last loaded element
    # driver.find_element(By.TAG_NAME, "body").send_keys(Keys.END)
    driver.execute_script(f"document.body.style.zoom = '1%'")

    time.sleep(2)  # Adjust the sleep time if needed

    # Find all elements with the specified class name
    elements = driver.find_elements(
        By.CLASS_NAME, "product-list_ProductList__item__LiiNI"
    )

    # If we have at least 200 elements, break the loop
    if len(elements) >= 200:
        break

    # Update the number of elements loaded
    prev_num_elements = len(elements)


```

```python
products = driver.find_elements(
    By.CLASS_NAME,
    "product-list_ProductList__item__LiiNI",
)

```

```python
product_links = []
for i, product in enumerate(products):
    print(f"element ({i}): --------------------------------")
    product_links.append(product.find_element(By.TAG_NAME, "a").get_attribute("href"))
    print(f"succesfully added the product link...")
    
```

```python
driver.get(product_links[0])
driver.implicitly_wait(5)
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
def get_detailed_specs(driver):
    detailed_specs = {}
    
    # Wait for the section to be visible
    try:
        section_locator = (By.XPATH, '//*[@id="__next"]/div[1]/div[3]/div[3]/div[2]/div[10]/div[2]/div[1]/section')
        WebDriverWait(driver, 10).until(EC.visibility_of_element_located(section_locator))
    except TimeoutException:
        
        return  
    # Click on the expandable object
    try:
        expandable_locator = (By.XPATH, '//*[@id="__next"]/div[1]/div[3]/div[3]/div[2]/div[10]/div[2]/div[1]/section/span')
        expandable_obj = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(expandable_locator))
    except TimeoutException:
        return   
    expanded = driver.execute_script("arguments[0].click();", expandable_obj);

    # Extract specifications
    try:
        detailed_specs["name"] = driver.find_element(By.XPATH, '//*[@id="__next"]/div[1]/div[3]/div[3]/div[2]/div[2]/div[2]/div[1]/div/div/nav/a[1]/span').text
    except NoSuchElementException:
        pass 

    try:
        detailed_specs["model"] = driver.find_element(By.XPATH, '//p[text()="مدل"]/following-sibling::div/p').text
    except NoSuchElementException:
        pass                                                        
    
    try:
        detailed_specs["price"] = driver.find_element(By.XPATH, '//span[@data-testid="price-no-discount"]').text
    except NoSuchElementException:
        pass
    
    try:
        annonce_date_locator = (By.XPATH, '//p[text()="زمان معرفی"]/following-sibling::div/p')
        annonce_date = WebDriverWait(driver, 10).until(EC.visibility_of_element_located(annonce_date_locator))
        detailed_specs["annonce_date"] = annonce_date.text
    except TimeoutException:
        pass
    except NoSuchElementException:
        pass
    
    try:
        detailed_specs["OS"] = driver.find_element(By.XPATH, '//*[@id="__next"]/div[1]/div[3]/div[3]/div[2]/div[10]/div[2]/div[1]/section/div[9]/div/div[1]/div/p').text
    except NoSuchElementException:
        pass
    
    try:
        detailed_specs["num_of_user_ratings"] = driver.find_element(By.XPATH, '//*[@id="__next"]/div[1]/div[3]/div[3]/div[2]/div[2]/div[2]/div[2]/div[1]/div[2]/div[1]/p[2]').text
    except NoSuchElementException:
        pass
    
    try:
        detailed_specs["users_rating"] = driver.find_element(By.XPATH, '//*[@id="__next"]/div[1]/div[3]/div[3]/div[2]/div[2]/div[2]/div[2]/div[1]/div[2]/div[1]/p[1]').text
    except NoSuchElementException:
        pass
    
    return detailed_specs
```

```python
mobiles = {}

for link in product_links:
    driver.get(link)
    driver.implicitly_wait(5)
    try:
        dets = get_detailed_specs(driver)
    except NoSuchElementException:
        pass
    print(dets)
    mobiles[link]=dets

mobiles
```
